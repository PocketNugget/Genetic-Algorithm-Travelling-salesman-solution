```python
class Node:
    def __init__(self, name, neighbors):
        self.name = name
        # A dictionary to store attributes of the node's neighbors
        self.neighbors = neighbors 
        
    def travel_time(self, node):
        # Returns the travel time to a specified neighboring node
        return self.neighbors[node.name]['time']
    
    def __repr__(self):
        # Returns a string representation of the node
        return "(" + str(self.name) + ")"
```

```python
def find_index(node: str) -> int:
    # Iterate through each index in the range of the length of nodeList
    for i in range(len(nodeList)):
        # Check if the name of the node at index i matches exactly with the given node
        if nodeList[i].name == node:
            # If found, return the index i
            return i
        # Check if the given node is a substring of the name of the node at index i
        elif node in nodeList[i].name:
            # If found, return the index i
            return i
    # If the node is not found in nodeList, return -1
    return -1

# Call find_index function to find the index of the start node ("Rosario")
start_node_index = find_index("Rosario")

# Call find_index function to find the index of the end node ("San Lazaro")
end_node_index = find_index("San Lazaro")

# Print the index of the start node and end node
print("Start Node Index:", start_node_index)
print("End Node Index:", end_node_index)
```

```python

def createGenome(nodeList):
    genome = []  
    remaining_nodes = nodeList.copy()
    current_node = remaining_nodes[start_node_index]
    remaining_nodes.remove(current_node)
    genome.append(current_node)

    while remaining_nodes:
        available_nodes = [node for node in remaining_nodes if node.name in current_node.neighbors]
        if not available_nodes:
            return None  # If no available nodes, return None indicating failure
        next_node = random.choice(available_nodes)  # Randomly select the next node from available nodes
        if next_node == nodeList[end_node_index]:
            genome.append(next_node)
            return genome  # If the end node is reached, return the genome
        else:
            genome.append(next_node)
            remaining_nodes.remove(next_node)  # Remove the selected node from remaining nodes
            current_node = next_node  # Update current node for next iteration

    return genome  # Return the constructed genome
```

```python
class Fitness:
    def __init__(self, genome):
        self.genome = genome  # Store the genome (sequence of nodes)
        self.time = 0  # Initialize the time attribute to zero
        self.fitness = 0.0  # Initialize the fitness attribute to zero

    def routeDistance(self):
        path_time = 0  # Initialize the path_time variable to zero
        for i in range(0, len(self.genome)):
            fromNode = self.genome[i]  # Get the current node from the genome
            if i + 1 < len(self.genome):
                toCity = self.genome[i + 1]  # Get the next node in the genome if exists
            if toCity.name in fromNode.neighbors:
                # If the next node's name is in the neighbors of the current node
                path_time += fromNode.travel_time(toCity)  # Add the travel time to path_time
        self.time = path_time  # Update the time attribute with the total travel time
        return self.time  # Return the total travel time

    def routeFitness(self):
        if self.fitness == 0:
            # If fitness is not calculated yet
            time = self.routeDistance()  # Calculate the route distance (time)
            self.fitness = 1 / float(time)  # Calculate fitness as reciprocal of time
        return self.fitness  # Return the fitness value
```

```python
def rankGenomes(population):
    fitnessResults = {}  # Initialize an empty dictionary to store fitness results
    # Calculate fitness for each genome in the population
    for i in range(0, len(population)):
        fitnessResults[i] = Fitness(population[i]).routeFitness()  # Calculate fitness using Fitness class

    # Sort fitness results in descending order based on fitness values
    sorted_results = sorted(fitnessResults.items(), key=operator.itemgetter(1), reverse=True)

    return sorted_results  # Return the sorted fitness results
```

```python
def selection(popRanked, eliteSize):
    selectionResults = []  # Initialize an empty list to store selected indices
    df = pd.DataFrame(np.array(popRanked), columns=["Index", "Fitness"])  # Create a DataFrame from popRanked
    df['cum_sum'] = df.Fitness.cumsum()  # Calculate cumulative sum of fitness values
    df['cum_perc'] = 100 * df.cum_sum / df.Fitness.sum()  # Calculate cumulative percentage of fitness

    # Select eliteSize number of top-ranked individuals directly
    for i in range(0, eliteSize):
        selectionResults.append(popRanked[i][0])  # Append index of top-ranked individual to selectionResults

    # Select the rest of the individuals based on roulette wheel selection
    for i in range(0, len(popRanked) - eliteSize):
        pick = 100 * random.random()  # Randomly pick a number between 0 and 100
        for i in range(0, len(popRanked)):
            if pick <= df.iat[i, 3]:  # Check if the random number falls within the cumulative percentage
                selectionResults.append(popRanked[i][0])  # Append index of selected individual to selectionResults
                break

    return selectionResults  # Return the list of selected indices (representing selected genomes)
```

```python
def matingPool(population, selectionResults):
    matingpool = []  # Initialize an empty list to store selected genomes (individuals)
    
    # Populate mating pool with genomes based on selection results
    for i in range(0, len(selectionResults)):
        index = selectionResults[i]  # Get the index of the selected genome from selectionResults
        matingpool.append(population[index])  # Append the selected genome to the mating pool
    
    return matingpool  # Return the list of selected genomes (mating pool)
```

```python
def mutate(individual, mutationRate):
    # Iterate through each gene (node) in the individual's genome
    for swapped in range(len(individual)):
        # Check if a random number is less than the mutation rate
        if random.random() < mutationRate:
            # Select another random gene to swap with
            swapWith = int(random.random() * len(individual))
            
            # Get the nodes (genes) at the current position and the swap position
            node1 = individual[swapped]
            node2 = individual[swapWith]
            
            # Check if the swap is valid based on neighbor constraints
            if (swapped == 0 or node2.name in individual[swapped - 1].neighbors) and \
               (swapped == len(individual) - 1 or node2.name in individual[swapped + 1].neighbors) and \
               (swapWith == 0 or node1.name in individual[swapWith - 1].neighbors) and \
               (swapWith == len(individual) - 1 or node1.name in individual[swapWith + 1].neighbors):
                
                # Perform the swap between the two genes (nodes)
                individual[swapped] = node2
                individual[swapWith] = node1
            else:
                continue  # If swap is not valid, continue to next iteration

    return individual  # Return the mutated individual's genome
```

```python
def mutatePopulation(population, mutationRate):
    mutatedPop = []  # Initialize an empty list to store mutated individuals (genomes)
    
    # Iterate through each individual in the population
    for ind in range(0, len(population)):
        # Mutate the individual's genome using the mutate function
        mutatedInd = mutate(population[ind], mutationRate)
        
        # Append the mutated individual to the mutated population
        mutatedPop.append(mutatedInd)
    
    return mutatedPop  # Return the mutated population (list of mutated individuals)
```
